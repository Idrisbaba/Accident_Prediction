### Description for Accident Prediction Dataset

The Data size and quality is very important to be able to allow us carr out machine learning models which is why I used the daily counts of road traffic accidents in London, along with environmental, temporal, and spatial information. The goal here is to predict the likelihood of accidents occurrence based on different variables such as weather conditions, time of the day, road type, location of accident, and other contributing factors. This dataset has been processed to facilitate analysis for supervised machine learning tasks.

Here is a list of the variables in the dataset:

- **Daily Count of Accidents**: The total number of accidents occurring per day. This count is used as the response in the regression task.
- **Indicator of Severity**: The most common severity of accidents per day (e.g., fatal, serious, or slight).
- **Indicator of Season**: Either spring, summer, fall, or winter, derived from the date.
- **Average Accident Timing**: The mean hour of accidents on a given day, representing temporal patterns.
- **Indicator Whether the Day Was a Weekend**: A binary indicator to flag weekends versus weekdays.
- **Weather Conditions**: Categories describing weather conditions during the day, such as:
  - **'GOOD'**: Fine no high winds, clear, or few clouds.
  - **'FOG/MIST'**: Foggy or misty weather.
  - **'RAIN/SNOW/STORM'**: Conditions involving rain, snow, or storms.
- **Spatial Features**: Longitude and latitude aggregated into clusters to identify accident-prone areas in London.
- **Road Design**: Features like `road type or junction information that describe the infrastructure associated with accidents.

---

```python
# Check for missing values
missing_values = data.isnull().sum()

# Filter to show only columns with missing values
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

if not missing_values.empty:
    print("⚠️ Columns with missing values:")
    print(missing_values)
else:
    print("✅ No missing values detected in the dataset.")
```

```python
# Fill missing spatial values but flag them
data['longitude_missing'] = data['longitude'].isnull().astype(int)
data['latitude_missing'] = data['latitude'].isnull().astype(int)

# Now fill them (temporarily using the mean as you did)
data['longitude'] = data['longitude'].fillna(data['longitude'].mean())
data['latitude'] = data['latitude'].fillna(data['latitude'].mean())

# Confirm no missing values
print("✅ Missing values after handling:")
print(data[['longitude', 'latitude']].isnull().sum())
```

```python
# Make sure day_of_week is numeric and handle unexpected values
data['day_of_week'] = pd.to_numeric(data['day_of_week'], errors='coerce')

# Create weekend indicator: 1 = weekend (Saturday=7, Sunday=1), 0 = weekday
data['weekend'] = data['day_of_week'].apply(lambda x: 1 if x in [1, 7] else 0)

# Optional: map numeric day to actual names (for visualization or readability)
day_map = {
    1: 'Sunday', 2: 'Monday', 3: 'Tuesday', 4: 'Wednesday',
    5: 'Thursday', 6: 'Friday', 7: 'Saturday'
}
data['day_name'] = data['day_of_week'].map(day_map)

# Verify the feature
print(data[['day_of_week', 'day_name', 'weekend']].head(10))
```

```python
# Convert 'time' column to datetime format safely
data['time_parsed'] = pd.to_datetime(data['time'], format='%H:%M', errors='coerce')

# Extract the hour (handles NaT safely)
data['hour'] = data['time_parsed'].dt.hour

# Define time bucket categorization
def categorize_time(hour):
    if pd.isnull(hour):  # Catch missing or invalid time
        return 'Unknown'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour <= 23:
        return 'Evening'
    else:
        return 'Midnight'  # 0 <= hour < 6

# Apply the function to categorize each row
data['time_bucket'] = data['hour'].apply(categorize_time)

# Preview
print(data[['time', 'hour', 'time_bucket']].head())
```

```python
# Check missing values
print("Missing values in key columns:")
print(merged_gdf[['weather_conditions', 'light_conditions', 'road_type']].isnull().sum())

# Impute missing values
merged_gdf['weather_conditions'] = merged_gdf['weather_conditions'].replace(-1, np.nan).fillna(merged_gdf['weather_conditions'].mode()[0])
merged_gdf['light_conditions'] = merged_gdf['light_conditions'].replace(-1, np.nan).fillna(merged_gdf['light_conditions'].mode()[0])
merged_gdf['road_type'] = merged_gdf['road_type'].replace(-1, 9)  # Treat "-1" as "Unknown"
```

```python
# Ordinal encoding for road_type (adjust order if needed)
road_type_mapping = {
    1: 0,  # Roundabout
    2: 1,  # One way street
    3: 2,  # Dual carriageway
    6: 3,  # Single carriageway
    7: 4,  # Slip road
    9: 5,  # Unknown
    12: 6, # One way street/Slip road
    -1: 5  # Treat "-1" as "Unknown"
}
merged_gdf['road_type_encoded'] = merged_gdf['road_type'].map(road_type_mapping)

# One-hot encode weather and light conditions
weather_dummies = pd.get_dummies(merged_gdf['weather_conditions'], prefix='weather')
light_dummies = pd.get_dummies(merged_gdf['light_conditions'], prefix='light')
merged_gdf = pd.concat([merged_gdf, weather_dummies, light_dummies], axis=1)
```

```python
# Define features (now including accident_count)
features = [
    'hour', 'road_type_encoded', 'speed_limit', 'accident_count',
    'urban_or_rural_area', 'weekend',
    'weather_1', 'weather_2', 'weather_4', 'weather_5', 'weather_7', 'weather_8', 'weather_9',
    'light_1', 'light_4', 'light_5', 'light_6', 'light_7'
]
```

```python
# After merging accident_count and defining features, train the model:

# Define target
target = 'accident_severity'

# Split data into features and target
X = merged_gdf[features]
y = merged_gdf[target]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3, 
    stratify=y, 
    random_state=42
)

# Initialize and train the model
model = RandomForestClassifier(
    n_estimators=200,
    class_weight={1: 10.0, 2: 5.0, 3: 1.0},  # Adjust weights as needed
    random_state=42
)
model.fit(X_train, y_train)

# Now evaluate
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['Fatal', 'Serious', 'Slight']))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import silhouette_score

# Apply KMeans clustering to spatial coordinates
spatial_data = data[['longitude', 'latitude']].dropna()  # Handle missing data if present
kmeans = KMeans(n_clusters=5, random_state=42)
data['accident_cluster'] = kmeans.fit_predict(spatial_data)

# Evaluate clustering performance with Silhouette Score
silhouette_avg = silhouette_score(spatial_data, data['accident_cluster'])
print(f"Silhouette Score: {silhouette_avg:.2f}")  # Indicates clustering quality (closer to 1 is better)

# Visualize clusters
plt.figure(figsize=(12, 8))
plt.scatter(data['longitude'], data['latitude'], c=data['accident_cluster'], cmap='viridis', s=10, alpha=0.6)
plt.colorbar(label='Cluster')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('KMeans Clustering of Accident Locations', fontsize=14)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Optional: Export clustered data for further analysis
data.to_csv('clustered_accidents.csv', index=False)
```

```python
cart_default = DecisionTreeRegressor(random_state=0)
cart_default.fit(train_x, train_y)

print(f"Tree depth: {cart_default.get_depth()}")

# Hyperparameter tuning
hyperparameters = {'max_depth': [10, 20, 30], 'min_samples_split': [2, 4, 6]}
clf = GridSearchCV(cart_default, hyperparameters, cv=5)
clf.fit(train_x, train_y)

print(f"Best parameters: {clf.best_params_}")
print(f"Best score: {clf.best_score_}")

dt_final = DecisionTreeRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=0)
dt_final.fit(train_x, train_y)
```

```python
# Evaluate Decision Tree Model
print("Decision Tree R² on training data:", dt_final.score(train_x, train_y))
print("Decision Tree R² on testing data:", dt_final.score(test_x, test_y))

# Calculate RMSE
print("Decision Tree RMSE:", np.sqrt(mean_squared_error(test_y, dt_final.predict(test_x))))
```

```python
rf = RandomForestRegressor(random_state=0)
rf.fit(train_x, train_y)

# Hyperparameter tuning
hyperparameters_rf = {'max_depth': [10, 20, 30], 'n_estimators': [50, 100, 150]}
clf_rf = GridSearchCV(rf, hyperparameters_rf, cv=5)
clf_rf.fit(train_x, train_y)

print(f"Best parameters: {clf_rf.best_params_}")
rf_final = RandomForestRegressor(max_depth=clf_rf.best_params_['max_depth'], n_estimators=clf_rf.best_params_['n_estimators'], random_state=0)
rf_final.fit(train_x, train_y)
```

```python
# Compare R² scores across models
models = {'Decision Tree': dt_final, 'Random Forest': rf_final}
comparison = {name: [model.score(train_x, train_y), model.score(test_x, test_y)] for name, model in models.items()}

comparison_df = pd.DataFrame.from_dict(comparison, orient='index', columns=['R² Train', 'R² Test'])
print(comparison_df)
comparison_df.head()
```
